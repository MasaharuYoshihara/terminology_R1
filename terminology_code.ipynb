```python
# Anatomical terminology supported by FEDERATIVE INTERNATIONAL PROGRAM ON ANATOMICAL TERMINOLOGIES
# https://www4.unifr.ch/ifaa/Public/EntryPage/TA98%20Tree/Alpha/All%20KWIC%20EN.htm
# viewed 14 February, 2021

# This code is following a tutorial by Mate Pocs, 2020
# https://towardsdatascience.com/pdf-text-extraction-in-python-5b6ab9e92dd
# viewed 14 February, 2021

# file names in number are PMID

import pandas as pd
import PyPDF2
import data_func
import csv

from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English

nlp = English()

# terminology.txt is list of anatomical terminology 
# supported by FEDERATIVE INTERNATIONAL PROGRAM ON ANATOMICAL TERMINOLOGIES

with open("terminology.txt") as text_file:
    contents = text_file.read()
    TAdata = contents.split(' ')

# use the directory for your own   

reader = PyPDF2.PdfFileReader(
    'miniconda3/envs/terminology/original/29067619.pdf')

#print(reader.documentInfo)

num_of_pages = reader.numPages
#print('Number of pages: ' + str(num_of_pages))

writer = PyPDF2.PdfFileWriter()

# set page number for each article

for page in range(0,5):

    writer.addPage(reader.getPage(page))
    
output_filename = 'miniconda3/envs/terminology/modified/29067619_modified.pdf'

with open(output_filename, 'wb') as output:
    writer.write(output)
    
text = data_func.convert_pdf_to_string(
    'miniconda3/envs/terminology/modified/29067619_modified.pdf')

text = text.replace('.','')
text = text.replace('/n','')
text = text.replace('\n','')
text = text.replace('\xa0','')
text = text.split(' ')

data = []
for word in text:
    lexeme = nlp.vocab[word]
    if lexeme.is_stop == False:
        data.append(word)

match = set(data) & set(TAdata)
match = str(match)

with open('result/29067619.txt','w') as f:
    f.write(match)
    f.close()
```

```python
import os
import glob
import collections

summary = []

# use the directory for your own

for filename in glob.glob('result/*.txt'):
    with open(os.path.join('', filename), 'r') as f:
        contents = f.read()
        contents = contents.replace("{", "")
        contents = contents.replace("}", "")
        summary.append(contents)
        f.close

summary = str(summary)
summary = summary.replace('"','')
summary = summary.replace("'",'')
summary = summary.replace(',','')
summary = summary.replace('[','')
summary = summary.replace(']','')
summary = summary.split(" ")
mycount = collections.Counter(summary)
#print(mycount)

mycount = str(mycount)

# make a summary file

with open('Version20210215.txt','w') as f:
    f.write(mycount)
    f.close()

PMIDs = glob.glob('result/*.txt')
PMIDs = str(PMIDs)
PMIDs = PMIDs.replace(".txt', 'result/",' ')
PMIDs = PMIDs.replace("'result/","")
PMIDs = PMIDs.replace(".txt'","")
print(PMIDs)
```
